### Data importing and pre-processing

```python
import pandas as pd
df = pd.read_csv('/content/Customer Survey.csv')
```

```python
print(df.shape)
df.head()
```

```python
df_ntime = df.dropna().reset_index(drop=True)
df_ntime = df.drop('Transaction', axis=1)
df_ntime = df_ntime.replace('ไม่เคย', 0)
df_ntime = df_ntime.replace('เคย', 1)
df_ntime.head(5)
```

```python
print(df.info())
```

```python
print(df_ntime.info())
```

```python
data_ibs = pd.DataFrame(index=df_ntime.columns,columns=df_ntime.columns)
data_ibs
```

```python
for i in range(0,len(data_ibs.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(data_ibs.columns)) :
      # Fill in placeholder with cosine similarities
      data_ibs.iloc[i,j] = 1-cosine(df_ntime.iloc[:,i],df_ntime.iloc[:,j])
data_ibs
```

```python
links = data_ibs.stack().reset_index()
links.columns = ['item1', 'item2','cosine_value']
links
```

```python
links_filtered=links.loc[ (links['cosine_value'] > 0.80) & (links['item1'] != links['item2']) ]
links_filtered
```

##Top 5 Cosine Similarity


```python
data_similarity = pd.DataFrame(index=data_ibs.columns,columns=range(1,6))

for i in range(0,len(data_ibs.columns)):
    data_similarity.iloc[i,:5]=data_ibs.iloc[0:,i].sort_values(ascending=False)[:5].index
    
data_similarity.head(5)
```
